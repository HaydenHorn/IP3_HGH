# 5.Remove all the symbols (\money sign\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip). The data frame will be look like after that:

```python
car_sales['Price'] = car_sales['Price'].str.replace(',', '')
car_sales['Price'] = car_sales['Price'].str.replace('$', ' ')
car_sales.head()
```

# 6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales['Prices']=car_sales['Price'].astype('int64')
```

```python
car_sales["Total Sales"] = car_sales['Price'].cumsum()
```
